```python
import csv
import pandas as pd
```

```python
csvFile = '..\Dados\conventional_weather_stations_inmet_brazil_1961_2019.csv'
csvNewFile = 'temperaturasJP.csv'
```

```python
df1 = {}
df1 = pd.read_csv(csvFile, sep=';')
```

```python
df1.head()
```

```python
isJP = df1['Estacao'] == 82798
```

```python
dfJP = df1[isJP].copy()
```

```python
dfJP.head()
```

```python
dfJP['Data'] = pd.to_datetime(dfJP.Data)
```

```python
dfJP['Data'] = dfJP['Data'].dt.strftime('%Y-%m')
```

```python
dfJP['Data'].head()
```

```python
gpbJP = dfJP.groupby('Data', as_index=True).agg({"TempMinima":"min","TempMaxima":"max"})
```

```python
resultado = gpbJP.dropna(axis=0, how='any').copy()
```

```python
resultado.to_csv(csvNewFile)
```
